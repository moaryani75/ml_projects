```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import minmax_scale
from sklearn.cluster import KMeans ,DBSCAN ,MeanShift 
```

```python
# prepration
x ,y =make_blobs(2000 ,centers =4 ,cluster_std=0.6)
x =minmax_scale(x ,feature_range=(0,1)).astype(np.float32)
y =y.reshape(-1,1).astype(np.int32)
plt.style.use('dark_background')
plt.figure(num =1 ,figsize =(7,5) ,dpi =100) 
plt.scatter(x[:,0] ,x[:,1] ,s =10 ,c ='b' ,marker ='o')
plt.xticks([]);plt.yticks([])
plt.xlabel('f_1' ,{'size' :12 ,'weight' :'bold'} ,10)
plt.ylabel('f_2' ,{'size' :12 ,'weight' :'bold'} ,10)
plt.title('cluster_plot' ,{'size' :18 ,'weight' :'bold'} ,pad =10)
plt.show()
```

```python
# details
# -------
```

```python
# K-means_model
# -------------
k_means =KMeans(n_clusters =4)
# training_and_prediction
labels_kmeans =k_means.fit_predict(x).reshape(-1,1)
centers_kmeans =k_means.cluster_centers_
```

```python
# dbscan_model
# ------------
dbscan =DBSCAN(eps =0.07 ,min_samples =50)
# training
dbscan.fit(x)
# prediction
labels_dbscan =dbscan.labels_.reshape(-1,1)
```

```python
# mean_shift_model
# ----------------
mean_shift =MeanShift(bandwidth =0.07) 
# training
mean_shift.fit(x) 
# prediction
labels_mean_shift =mean_shift.predict(x).reshape(-1,1)
centers_mean_shift =mean_shift.cluster_centers_
```
